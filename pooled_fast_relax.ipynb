```python
import os
## WT complex
script= "/home/mhoum/drive1/fastrelax/pooled_fast_relax.py"

wt_complex_pdb= "/home/mhoum/drive1/fastrelax/input_structures/complex/1brs_complex_equilibrated.pdb"
output_folder= "/home/mhoum/drive1/fastrelax/output_conformers/complex/wt"

n_relax= 12
n_processes= 2
n_cycles= 100


# Run parallel relaxation
cmd= f"""conda init;
conda activate pyrosetta_env;
python {script} -i {wt_complex_pdb} -o {output_folder} -n {n_relax} -p {n_processes} --cycles {n_cycles}"""

print(cmd)
#os.system(cmd)

```

```python
import pyrosetta

## pyrosetta/database/chemical/residue_type_sets/fa_standard/merge_residue_behaviors.txt was modified to prevent merging the ACE and NME residues with the standard amino acids

# Initialize PyRosetta with additional residue type sets
#opts = "-use_terminal_residues true"
pyrosetta.init(options="-make_rot_lib:use_terminal_residues true -maintain_links -no_optH")

# Load your structure
pose = pyrosetta.pose_from_pdb(wt_complex_pdb)

pose.dump_pdb("check_format2.pdb")
```
